Сортування вставлянням має асимптотичну складність O(n^2), оскільки для кожного елемента масиву може виконуватися до n порівнянь і зсувів.
Сортування бульбашкою також має складність O(n^2), тому що використовує два вкладені цикли, які проходять по всьому масиву.
У найгіршому випадку кількість операцій пропорційна n².

```python
import random
import time
import matplotlib.pyplot as plt

def insertion_sort(nums):
    for j in range(1, len(nums)):
        key = nums[j]
        i = j - 1
        while i >= 0 and nums[i] > key:
            nums[i + 1] = nums[i]
            i -= 1
        nums[i + 1] = key


def bubble_sort(nums):
    n = len(nums)
    for i in range(n):
        for j in range(0, n - i - 1):
            if nums[j] > nums[j + 1]:
                nums[j], nums[j + 1] = nums[j + 1], nums[j]

sizes = [5, 10, 50, 100, 500, 1000, 2000, 3000, 4000, 5000]

times_insert = []
times_bubble = []

for n in sizes:
    arr = [random.randint(0, 10000) for _ in range(n)]

    arr_copy1 = arr.copy()
    arr_copy2 = arr.copy()

    start = time.time()
    insertion_sort(arr_copy1)
    times_insert.append(time.time() - start)

    start = time.time()
    bubble_sort(arr_copy2)
    times_bubble.append(time.time() - start)

plt.plot(sizes, times_insert, label="Insertion Sort")
plt.plot(sizes, times_bubble, label="Bubble Sort")

plt.xlabel("Розмір масиву (n)")
plt.ylabel("Час (секунди)")
plt.title("Порівняння алгоритмів сортування")
plt.legend()

plt.show()
```

## Висновок

Обидва алгоритми мають складність O(n^2), але на практиці сортування вставлянням працює швидше за бульбашкове сортування.
Зі збільшенням розміру масиву час виконання значно зростає.

## Контрольні питання

### 1. Що таке O-нотація і чим вона відрізняється від Ω-нотації?

O-нотація (Big-O) описує асимптотичну верхню межу складності алгоритму, тобто показує найгірший випадок зростання часу виконання.
Ω-нотація (Big-Omega) описує асимптотичну нижню межу складності алгоритму, тобто показує найкращий можливий випадок.
Основна різниця: O-нотація — це верхня оцінка, Ω-нотація — нижня.

### 2. Яку часову складність має алгоритм сортування методом вставляння за найсприятливіших умов?

У найкращому випадку масив вже відсортований.  
Тоді кожен елемент перевіряється лише один раз без зсувів.
Отже, алгоритм виконує лінійний прохід по масиву.
Часова складність: **O(n)**.

### 3. Визначення ефективного алгоритму

Ефективний алгоритм — це алгоритм, який виконує задачу з мінімальними витратами часу та пам’яті, тобто має оптимальну асимптотичну складність для даної задачі.

### 4. Поняття головного параметра (розміру) задачі

Головний параметр задачі — це величина, яка визначає вхідні дані алгоритму і впливає на його складність.
Зазвичай це:
- кількість елементів масиву (n)
- розмір графа (вершини/ребра)
- довжина рядка тощо
