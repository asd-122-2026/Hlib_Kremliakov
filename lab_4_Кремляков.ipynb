```python
# факторіал (ітеративний)
def FacSimple(n):
    factorial = 1
    i = 1
    while i <= n:
        factorial *= i
        i += 1
    return factorial

print(FacSimple(5))
```

```python
# факторіал (рекурсія)
def fac(n):
    if n == 0:
        return 1
    return fac(n - 1) * n

print(fac(5))
```

```python
# fibonacci (рекурсія)
def fibonacci(n):
    if n == 0:
        return 0
    if n in (1, 2):
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(10))
```

```python
# merge sort
def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result += left[i:]
    result += right[j:]
    return result


def mergesort(lst):
    if len(lst) < 2:
        return lst

    middle = len(lst) // 2
    left = mergesort(lst[:middle])
    right = mergesort(lst[middle:])
    return merge(left, right)


a = [6, 5, 3, 1, 8, 7, 2, 4]
print(mergesort(a))
```

```python
def factorial_for(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print(factorial_for(5))
```

## Складність факторіалу

Алгоритм обчислення факторіалу через цикл for має часову складність O(n), оскільки виконує один прохід по всіх значеннях від 1 до n.
Рекурсивний варіант також має складність O(n), оскільки виконується n рекурсивних викликів.

## Складність Фібоначчі

Рекурсивний алгоритм обчислення чисел Фібоначчі має експоненційну складність O(2^n), оскільки кожен виклик породжує два нових виклики.

## Складність сортування злиттям

Алгоритм merge sort має рекурентну формулу:
T(n) = 2T(n/2) + O(n)

За головною теоремою рекурсії:
a = 2, b = 2, d = 1
log₂2 = 1 → d = log_b(a)

Отже, складність:
O(n log n)

## Контрольні питання

1. Ω-нотація
Ω-нотація описує нижню межу складності алгоритму, тобто найкращий випадок його роботи.

3.

→ O(1.01^N)

4.

5. Розпаралелювання
Ідея розпаралелювання полягає у виконанні декількох частин алгоритму одночасно для зменшення загального часу виконання.
У цій роботі можливе розпаралелювання для merge sort, оскільки підзадачі (ліва і права частина) незалежні.
